## Step 3 — Data Cleaning

```python
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
print(df['Total Charges'].isnull().sum())
```

```python
df.dropna(subset=['Total Charges'], inplace=True)
print("Rows after dropping nulls:", df.shape[0])
```

```python
df['Churn Reason'] = df['Churn Reason'].fillna('Not Churned')
print(df['Churn Reason'].value_counts())
```

```python
cols_to_drop = ['Count', 'Lat Long', 'Latitude', 'Longitude', 'Zip Code', 'Country', 'State', 'City']
df.drop(columns=cols_to_drop, inplace=True)
print("Dropped columns. New shape:", df.shape)
```

```python
def tenure_group(tenure):
    if tenure <= 12:
        return '0-12 Months'
    elif tenure <= 24:
        return '13-24 Months'
    elif tenure <= 48:
        return '25-48 Months'
    else:
        return '49+ Months'

df['Tenure Group'] = df['Tenure Months'].apply(tenure_group)
print(df['Tenure Group'].value_counts())
```

```python
def churn_risk(score):
    if score >= 75:
        return 'High Risk'
    elif score >= 50:
        return 'Medium Risk'
    else:
        return 'Low Risk'

df['Churn Risk'] = df['Churn Score'].apply(churn_risk)
print(df['Churn Risk'].value_counts())
```

```python
print("Shape:", df.shape)
print("\nMissing Values:")
print(df.isnull().sum()[df.isnull().sum() > 0])
print("\nNew Columns Added:")
print(df[['Tenure Group', 'Churn Risk']].head())
```
